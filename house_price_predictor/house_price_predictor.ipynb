```python
# we replace he value with mean of the range
def convert_range(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft']=data['total_sqft'].apply(convert_range)
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby("location"):
        mean = np.mean(subdf.price_per_sqft)
        sd = np.std(subdf.price_per_sqft)
        diff = subdf[(subdf.price_per_sqft > (mean - sd)) & (subdf.price_per_sqft <= (mean + sd))]
        df_output = pd.concat([df_output, diff], ignore_index=True)
    return df_output

# Apply the function
data = remove_outliers_sqft(data)

# Show updated summary
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices, 
                    bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values
                )

    return df.drop(exclude_indices, axis='index')
```

```python
data = bhk_outlier_remover(data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pickle

# Load cleaned data
data = pd.read_csv("cleaned_data.csv")
X = data.drop(columns=['price'])
y = data['price']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define transformers
categorical = ['location']
numerical = ['total_sqft', 'bath', 'bhk']

column_trans = ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical),
    ('scaler', StandardScaler(), numerical)
])

# Create pipeline
pipe = make_pipeline(column_trans, Ridge())

# Train the model
pipe.fit(X_train, y_train)

# Save the model
pickle.dump(pipe, open('ridgemodel.pkl', 'wb'))
```
